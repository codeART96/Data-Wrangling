```python

def set_scores(row_items, number, data):
    '''Returns a count'''
    count = 0
    for column in data.columns:
        breakdown = row_items[column]
        if type(breakdown) == str:
            tup_break = breakdown[1]
            result = int(breakdown[4])
            if tup_break == number:
                count += result
    return count
```

```python
new_data['Extraversion'] = new_data.apply(lambda row_items : set_scores(row_items,'1', new_data), axis = 1)
new_data['Agreeableness'] = new_data.apply(lambda row_items : set_scores(row_items,'2', new_data), axis = 1)
new_data['Conscientiousness'] = new_data.apply(lambda row_items : set_scores(row_items,'3', new_data), axis = 1)
new_data['Neuroticism'] = new_data.apply(lambda row_items : set_scores(row_items,'4', new_data), axis = 1)
new_data['Imagination'] = new_data.apply(lambda row_items : set_scores(row_items,'5', new_data), axis = 1)
```

```python
def health_rate(score,score_name):
    '''A function that gives score for health risk rating'''

    if score <= 30:
        return 'high_risk'
    else:
        return 'low_risk'
    
joined_df['Neuroticism_score'] = joined_df['Neuroticism'].apply(lambda column: health_rate(column,'Neuroticism'))
joined_df['Agreeableness_score'] = joined_df['Agreeableness'].apply(lambda column: health_rate(column,'Agreeableness'))
joined_df['Conscientiousnessn_score'] = joined_df['Conscientiousness'].apply(lambda column: health_rate(column,'Conscientiousness'))
```
